```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_Loaned.csv')
```

```python
df.drop(columns={'Loan_ID'},axis=1,inplace=True)
```

```python
df.columns
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Gender'],hue='Loan_Status',data=df)
plt.title('Gender vs Loan Approval')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Married'],hue='Loan_Status',data=df)
plt.title('Married vs Loan Approval')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Dependents'],hue='Loan_Status',data=df)
plt.title('Dependents vs Loan Approval')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Education'],hue='Loan_Status',data=df)
plt.title('Education vs Loan Approval')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Self_Employed'],hue='Loan_Status',data=df)
plt.title('Self Employed vs Loan Approval')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x=df['Property_Area'],hue='Loan_Status',data=df)
plt.title('Property Area vs Loan Approval')
```

<p>
Gender: Males apply and get approved more than females.

Married: Married applicants have higher approval rates.

Dependents: Fewer dependents tend to correlate with higher loan approvals.

Education: Graduates have better loan approval chances.

Property Area: Semiurban residents have the highest approval rates, followed by urban and rural.

Self Employed: Non-self-employed applicants are more likely to get approved.
    
</p>

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=df['ApplicantIncome'],hue='Loan_Status',data=df)
plt.title('ApplicantIncome vs Loan_Status')
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=df['CoapplicantIncome'],hue='Loan_Status',data=df)
plt.title('coapplicantIncome vs Loan_Status')
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=df['LoanAmount'],hue='Loan_Status',data=df)
plt.title('Loan Amount vs Loan_Status')
```

<p>Here’s what your boxplots (feature vs. Loan_Status) show:

ApplicantIncome vs Loan_Status:
The distribution of applicant incomes is similar for both approved (Y) and not approved (N) loans. Most incomes are concentrated in a low range for both, with several high-income outliers in each group. High income alone does not guarantee loan approval.

CoapplicantIncome vs Loan_Status:
Like applicant income, coapplicant incomes for both approval statuses are low for most applicants, with a few high-income outliers. Loan approval does not significantly vary with coapplicant income.

LoanAmount vs Loan_Status:
Loan amounts are distributed similarly for both approved and unapproved loans. The majority are in the lower range, with some high-value outliers in both groups. Higher loan amounts don’t necessarily reduce approval chances.</p>

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x = df['ApplicantIncome'],y = df['LoanAmount'],hue='Loan_Status',data=df)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x = df['CoapplicantIncome'],y = df['LoanAmount'],hue='Loan_Status',data=df)
```

<p>
    There is no strong or very clear separation between approved and not approved loans based on applicant or coapplicant income versus requested loan amount. Both ‘Y’ and ‘N’ statuses are scattered similarly, suggesting that loan approval is influenced by other factors, not just income and loan amount.
</p>
