```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('Encoded_Loan_Data.csv')
```

```python
X = df[['Gender_Encode','Married_Encode','Education_Encode','Self_Employed_Encode', 'Dependents', 'ApplicantIncome',
       'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History','Property_Area_Encode']]

y = df['Loan_Status_Encode']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
params = {'max_depth':range(1,7),'min_samples_leaf':range(1,6),'n_estimators':range(25,201,25)}
```

```python
grid_rfc = GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,cv=5,verbose=2,n_jobs=3)
grid_rfc.fit(X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
grid_rfc.best_estimator_
```

```python
print('Score of Random Forest Algorithm is',round((grid_rfc.best_score_) * 100,2))
```

```python
y_pred = grid_rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print('Accuracy score on test data',round(accuracy_score(y_test,y_pred)*100,2))
print('Precision score on test data',round(precision_score(y_test,y_pred)*100,2))
print('Recall score on test data',round(recall_score(y_test,y_pred)*100,2))
print('F1 score on test data',round(f1_score(y_test,y_pred)*100,2))
```
