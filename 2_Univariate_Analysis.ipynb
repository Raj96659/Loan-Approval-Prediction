```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_Loaned.csv')
```

```python
df.drop(columns={'Loan_ID'},axis=1,inplace=True)
```

```python
df.columns
```

<h3>Count Plots</h3>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Gender',data=df)
plt.title('Male vs Female')
```

<p>Gender: Most loan applicants are male. Females are much fewer.</p>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Married',data=df)
plt.title('Married vs Not Married')
```

<p>Married: More applicants are married than not married.</p>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Education',data=df)
plt.title('Graduate vs Not Graduate')
```

<p>Education: Most are graduates. Not graduates are much fewer.</p>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Self_Employed',data=df)
plt.title('Self_Employed vs Not')
```

<p>Self_Employed: Most applicants are not self-employed. Self-employed applicants are a minority.</p>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Property_Area',data=df)
plt.title('Urban vs Rural vs Semi')
```

<p>Property_Area: Applicants are spread fairly evenly across urban, rural, and semi-urban areas.</p>

```python
plt.figure(figsize=(3,2))
sns.countplot(x='Loan_Status',data=df)
plt.title('Yes vs No')
```

<p>Loan_Status: More loans were approved (Y) than denied (N).</p>

<p>Findings</p>
<p>Most applicants are male, married, educated graduates, not self-employed, and their properties are in different areas. The majoritys loans were approved. </p>

<h2>Understanding the distribution</h2>

```python
plt.figure(figsize=(4,3))
sns.histplot(df['ApplicantIncome'])
```

```python
plt.figure(figsize=(4,3))
sns.histplot(df['CoapplicantIncome'])
```

```python
plt.figure(figsize=(4,3))
sns.histplot(df['LoanAmount'])
```

<p>ApplicantIncome: Most applicants have incomes below 10,000, with a large peak at the lower values. Very few applicants have extremely high incomes, visible as isolated bars far to the right.

CoapplicantIncome: Similar pattern: Most coapplicants have low incomes (below 5,000), with some very high outliers.

LoanAmount: Most loan amounts are between 100 and 200. There are some higher amounts (above 300), but these are much less common.</p>

<h3>Outlier Detection</h3>

```python
plt.figure(figsize=(4,3))
sns.boxplot(x = df['ApplicantIncome'])
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(x = df['CoapplicantIncome'])
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(x = df['LoanAmount'])
```

<p>ApplicantIncome: Most applicant incomes are clustered at lower values. There are many high-value outliers (very high incomes) that stand out far from the majority.

CoapplicantIncome: Most co-applicant incomes are also on the lower side, but there are very high outliers visible, just like in ApplicantIncome.

LoanAmount: Typical loan amounts are below 200, but several applicants requested much larger loans, which are visible as outliers.</p>
